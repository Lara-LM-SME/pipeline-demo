```python
import pandas as pd
from sklearn.pipeline import make_pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC

# Load training data
df = pd.read_csv("data/reviews.csv")  # needs columns: text, sentiment

# Build & fit the same model as train.py
C = 1.0
pipe = make_pipeline(
    TfidfVectorizer(ngram_range=(1,2), lowercase=True, min_df=1),
    LinearSVC(C=C, random_state=42)
)
pipe.fit(df["text"], df["sentiment"])

def predict_sentiment(s: str) -> str:
    return "positive" if pipe.predict([s])[0] == 1 else "negative"

# Try it eg "the light worked well", or "I didn't like it". Reflect on how you could improve the results for the second example.
example = "I didn't like it"
print(example, "→", predict_sentiment(example))
```
